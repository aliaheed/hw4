```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head(20)

school_data_complete.head(20)
#student_data.head(20)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school = school_data_complete["school_name"].nunique()

total_student = school_data_complete["student_name"].count()

total_budget = school_data["budget"].sum()

avg_math_score = school_data_complete["math_score"].mean()

avg_reading_score = school_data_complete["reading_score"].mean()

overall_passing_rate = (avg_math_score + avg_reading_score) / 2

students_m_score_over_70 = len(school_data_complete.loc[(school_data_complete["math_score"]>=70)])

percentage_passing_math = (students_m_score_over_70 / total_student) * 100

students_r_score_over_70 = len(school_data_complete.loc[(school_data_complete["reading_score"]>=70)])

percentage_passing_reading = (students_r_score_over_70 / total_student) * 100

summary_df = pd.DataFrame({"Total Schools": [total_school],
                           "Total Students": [total_student],
                           "Total Budget": [total_budget],
                           "Average Math Score": [avg_math_score],
                           "Average Reading Score": [avg_reading_score],
                           "% Passing Math": [percentage_passing_math],
                           "% Passing Reading": [percentage_passing_reading],
                           "% Overall Passing Rate": [overall_passing_rate]})

summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create a group based on the values in the 'maker' column


new_school_data=school_data[['school_name', 'budget']]

total_students = school_data_complete.groupby(['school_name','type'])['Student ID'].count()


total_students_df = total_students.to_frame().reset_index()
total_students_df = total_students_df.rename(columns={"type":"School Type", "Student ID":"Total Students"})


combine_total_school = pd.merge(new_school_data,total_students_df, on="school_name")

## -- Per Student Budget -- ##
combine_total_school["Per Student Budget"] = combine_total_school["budget"] / combine_total_school["Total Students"]
combine_total_school.head(20)

## -- Average Math Score -- ##
avg_math_score = school_data_complete.groupby(['school_name'])['math_score'].mean().rename("Average Math Score")
avg_math_score_df = avg_math_score.to_frame().reset_index()

combine_math_score = pd.merge(combine_total_school,avg_math_score_df, on="school_name")

## -- Average Reading Score -- ##
avg_reading_score = school_data_complete.groupby(['school_name'])['reading_score'].mean().rename("Average Reading Score")
avg_reading_score_df = avg_reading_score.to_frame().reset_index()


combine_total_school = pd.merge(combine_math_score,avg_reading_score_df, on="school_name")

## -- % Passing Math -- ##
school_math_score = school_data_complete[['school_name', 'math_score']]
school_math_score = school_math_score.loc[(school_math_score["math_score"]>=70)]
school_math_score = school_math_score.groupby(['school_name'])['math_score'].count().rename("count passed math")
school_math_score_df = school_math_score.to_frame().reset_index()

combine_total_school = pd.merge(combine_total_school,school_math_score_df, on="school_name")

combine_total_school["% Passing Math"] = (combine_total_school["count passed math"] / combine_total_school["Total Students"])*100

## -- % Passing Reading -- ##
school_reading_score = school_data_complete[['school_name', 'reading_score']]
school_reading_score = school_reading_score.loc[(school_reading_score["reading_score"]>=70)]
school_reading_score = school_reading_score.groupby(['school_name'])['reading_score'].count().rename("count passed reading")
school_reading_score_df = school_reading_score.to_frame().reset_index()

combine_total_school = pd.merge(combine_total_school,school_reading_score_df, on="school_name")

combine_total_school["% Passing Reading"] = (combine_total_school["count passed reading"] / combine_total_school["Total Students"])*100

## -- % Overall Passing Rate  -- ##
combine_total_school["% Overall Passing Rate"] = (combine_total_school["% Passing Math"] + combine_total_school["% Passing Reading"])/2
combine_total_school

combine_total_school = combine_total_school.rename(columns={"budget":"Total School Budget", "Student ID":"Total Students"})


combine_total_school = combine_total_school[["school_name", "School Type", "Total School Budget", "Per Student Budget",
                       "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]



combine_total_school["Total School Budget"] = combine_total_school["Total School Budget"].map("${:,.2f}".format)
combine_total_school["Per Student Budget"] = combine_total_school["Per Student Budget"].map("${:,.2f}".format)


combine_school_index = combine_total_school.set_index('school_name')
combine_school_index
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_performing_school = combine_school_index.sort_values(["% Overall Passing Rate"], ascending=False)
top_performing_school.reset_index(inplace=True)

top_performing_school=top_performing_school.nlargest(5,'% Overall Passing Rate')
top_performing_school

```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python

bottom_performing_school = combine_school_index.sort_values(["% Overall Passing Rate"], ascending=True)
bottom_performing_school.reset_index(inplace=True)

bottom_performing_school=bottom_performing_school.nsmallest(5,'% Overall Passing Rate')
bottom_performing_school
```
